```python
import pandas as pd
df = pd.read_csv(r'D:\task_folder\sources\IND_prep\data\loans\loans.csv',encoding='utf-8')
```

```python
df['non_decimal_feature']
```

```python
cols = 'non_decimal_feature'
```

```python
import numpy as np

def parse_non_decimal(val):
    try:
        if isinstance(val, str):
            val = val.lower().strip()
            if val.startswith('0b'):
                return int(val, 2)  # 2진수
            elif val.startswith('0x'):
                return int(val, 16)  # 16진수
            else:
                return int(val)  # 10진수 가정
        return val
    except (ValueError, TypeError):
        return np.nan
df[cols].apply(parse_non_decimal)
    
```

```python
df.head()
```

```python
from sentence_transformers import SentenceTransformer
# 새로운 함수: 문장형 → 벡터형
def sentence_to_vector(df):
    df = df.copy()
    cols = 'sentence_feature'
    if not cols or pd.isna(cols):
        return df
    model_name = 'sentence-transformers/all-MiniLM-L6-v2'
    model = SentenceTransformer(model_name)
    for col in cols:
        sentences = df[col].fillna('').tolist()
        embeddings = model.encode(sentences, show_progress_bar=False)
        # 벡터를 새로운 컬럼으로 추가
        for i in range(embeddings.shape[1]):
            df[f'{col}_vec_{i}'] = embeddings[:, i]
        df.drop(columns=[col], inplace=True)
    return df[cols]
```

```python
sentence_to_vector(df)
```
