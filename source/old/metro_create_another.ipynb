```python
def aa(df,a,b):
    df = df.copy()
    df.loc[(df['timestamp']>=a)&(df['timestamp']<=b), 'y'] = 1
    return df
```

```python
for i in range(0,41,2):
    df = aa(df,fault.iloc[i][0],fault.iloc[i+1][0])
```

```python
from imblearn.over_sampling import SMOTE
X = df.drop(columns=['timestamp', 'y'])
y = df['y']
oversample = SMOTE(sampling_strategy=0.05, random_state=0)
X, y = oversample.fit_resample(X, y)
df0 = pd.concat([X,y], 1)
df0['y'].value_counts(True)
```

```python
times = pd.date_range('2019-09-01', periods=df0.shape[0], freq='1min')
cols = ['timestamp'] + list(df0.columns)
df0['timestamp'] = times
df0 = df0[cols]
df0
```
