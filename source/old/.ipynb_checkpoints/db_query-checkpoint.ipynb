```python
# pip install SQLAlchemy  (미리 pip 설치 필요)

import pandas as pd
from sqlalchemy import create_engine

host='192.168.10.17'
database = 'jips'
user='root'
password='1234'
port=3306
```

```python
try:
    sql = "select * from POWER_METER_DAY where DT >= '2020-01-01' AND DT <= '2020-12-31'"
    engine = create_engine(f'mysql+mysqlconnector://{user}:{password}@{host}:{port}/{database}')    
    df = pd.read_sql(sql, con=engine)    
except Exception as e:
    print(e)
finally:
    if engine is not None:
        engine.dispose()
```
